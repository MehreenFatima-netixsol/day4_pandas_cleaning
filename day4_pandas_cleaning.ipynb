```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df = pd.read_csv(url)

df.head()
```

We first inspect the dataset to identify missing values. Understanding where missing data exists helps determine whether values should be removed or filled.

```python
df.isnull().sum()
```

Cabin has a large number of missing values and provides little value for this analysis, so it is removed. Age is filled using the median because it is a numerical feature and the median is less affected by outliers. Embarked has very few missing values and is filled using the mode (most common value).

```python
df = df.drop(columns=["Cabin"])
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

df.isnull().sum()
```

We inspect each column's data type to ensure numerical columns are stored as numeric values.

```python
df.dtypes
```

Duplicate records can introduce bias into analysis, so they are removed.

```python
print("Duplicates before:", df.duplicated().sum())

df = df.drop_duplicates()

print("Duplicates after:", df.duplicated().sum())
```

#Create Derived Column (Age Group)
Categorize passengers into Child, Adult, and Senior age groups.

```python
def age_group(age):
    if age < 18:
        return "Child"
    elif age < 60:
        return "Adult"
    else:
        return "Senior"

df["Age_Group"] = df["Age"].apply(age_group)

df.head()
```

Calculate ticket fare after adding a 10% service tax using a vectorized operation.

```python
df["Fare_With_Tax"] = df["Fare"] * 1.10

df.head()
```
